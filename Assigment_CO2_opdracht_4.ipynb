```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Graph 1**

```python
tables = pd.read_excel('./CO2 Emissies (bewerkt).xlsx',skiprows=3)
countries=tables
countries
```

```python
countries.columns = ['country', '1990', '2005', '2017', '2021','2017(% of world)', '2017 vs 1990: change (%)', 'per land area', 'per capita', 'total inc_lucf', 'total ex_lucf']
countries
```

```python
countries = countries.loc[3:]
countries = countries.drop([67])
countries
```

```python
from prompt_toolkit.formatted_text.utils import to_plain_text
top_5 = countries.loc[:,['country','1990','2005','2017','2021']]
top_5 = top_5.sort_values(by='2021',ascending=False).head(5)
top_5
```

```python
fig, ax =plt.subplots()
years = top_5.columns[1:]
for index, row in top_5.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title('Top 5 CO2 emitters')
plt.xlabel('Year')
plt.ylabel('Emission')
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.show()
```

**Graph 2**

```python
graph_2 = countries.loc[:,["country", "1990", "2005", "2017", "2021"]]
graph_2 = graph_2.loc[graph_2["1990"] >= 5]
graph_2
```

```python
pd.options.display.float_format = '{:.3f}'.format
#New colums with percentage of change per year based on 1990.
graph_2['1990vs1990ch'] = graph_2.loc[:,'1990'] / graph_2.loc[:,'1990'] * 100
graph_2['2005vs1990ch'] = graph_2.loc[:,'2005'] / graph_2.loc[:,'1990'] * 100
graph_2['2017vs1990ch'] = graph_2.loc[:,'2017'] / graph_2.loc[:,'1990'] * 100
graph_2['2021vs1990ch'] = graph_2.loc[:,'2021'] / graph_2.loc[:,'1990'] * 100
graph_2 = graph_2.loc[:,["country","1990vs1990ch", "2005vs1990ch", "2017vs1990ch", "2021vs1990ch"]] 
graph_2
graph_2_best = graph_2.sort_values(by='2021vs1990ch', ascending=True).head(3) 
graph_2_worst = graph_2.sort_values(by='2021vs1990ch', ascending=False).head(3) 
graph_2_best.columns = ["country", "1990", "2005", "2017", "2021"]
graph_2_best
```

```python
fig, ax =plt.subplots()
years = graph_2_best.columns[1:]
for index, row in graph_2_best.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title('Most percentage Of decrease changest in the world')
plt.xlabel('Year')
plt.ylabel('CO2 change in %')
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.show()
```

```python
graph_2_worst.columns = ["Country", "1990", "2005", "2017", "2021"]# Renaming for better use in graph.
graph_2_worst
```

```python
fig, ax =plt.subplots()
years = graph_2_worst.columns[1:]
for index, row in graph_2_worst.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title('Most percentage Of increase changest in the world')
plt.xlabel('Year')
plt.ylabel('CO2 change in %')
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.show()
```
